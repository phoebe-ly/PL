```python
import pandas as pd

# read csv file
dataa = pd.read_csv("File_14618.csv") 
```

```python
dataa #102年18歲以上人口目前嚼檳榔率-依縣市別分
```

```python
l1=list(dataa[(dataa["女Female(%)"]<1)]["依縣市別分by county"])#給女生小於1%的條件
l2=list(dataa[(dataa["男Male(%)"]<10)]["依縣市別分by county"])#給男生小於10%的條件
print("女生嚼檳榔人口沒超過1%的城市數量:", len(dataa[(dataa["女Female(%)"]<1)]))
print("男性嚼檳榔人口沒超過10%的城市數量:", len(dataa[(dataa["男Male(%)"]<10)]))
```

```python
s1=set(l1)
s2=set(l2)
print(s1)#女生嚼檳榔人口沒超過1%的城市
print(s2)#男性嚼檳榔人口沒超過10%的城市
```

```python
print(s1&s2)#(交集)女和男都滿足條件
print(s1|s2)#(聯集)女和男至少一個滿足條件
print(s1-s2)#(差集)女滿足條件，男不滿足
print(s2-s1)#(差集)男滿足條件，女不滿足
print(s1^s2)#(對稱差集)女和男只有一個滿足條件
```
