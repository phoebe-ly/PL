```python
#查詢威秀影城即將上映的電影資訊
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
url = "https://www.vscinemas.com.tw/vsweb/film/coming.aspx"#網址處
headers={
"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
}

res = requests.get(url,headers = headers)
print(res.text)
```

```python
#列出我要的資訊
soup = BeautifulSoup((requests.get(url)).text, "lxml")
movieList_items = soup.find_all('section', 'infoArea')

for item in movieList_items:
    name = item.find('h2').a.text.strip()
    english_name = item.find('h3').text.strip()
    release_time = item.find('time').text.strip()
     
    print('{}({})上映時間:{}'.format(name,english_name,release_time))
```

```python
#做成一個巢狀資料
soup = BeautifulSoup((requests.get(url)).text, "lxml")
movieList_items = soup.find_all('section', 'infoArea')
movie = {'name':[],'english_name':[],'release_time':[]}
for item in movieList_items:
    name = item.find('h2').a.text.strip()
    english_name = item.find('h3').text.strip()
    release_time = item.find('time').text.strip()
    movie.get('name').append([name])
    movie.get('english_name').append([english_name])
    movie.get('release_time').append([release_time])

    
movie
```

```python
#轉成 DataFrame
df = pd.DataFrame(movie)
print(df)
```

```python
#做成csv檔和下載
df.to_csv('movie.csv')
print("csv下載完成")
```

```python
#轉json檔和下載
with open('movie.json', 'w', encoding='utf-8') as file:
    json.dump(movie, file, ensure_ascii=False)  

print("json下載完成")
```
