```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
url1 = "http://www.meetpets.org.tw/pets/cat"
headers={
"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
}
res = requests.get(url1,headers = headers)
```

```python
#列出我要的資訊
soup1 = BeautifulSoup((requests.get(url1)).text, "lxml")
itemslist1 = soup1.find_all('div', 'view-item view-item-pets-cat')

p1 = []
for item in itemslist1:
    picture1 = item.find('div', 'field-item').img
    p1.append(picture1)
p1
```

```python
l1 = {"名字":[],"簡介":[]}
for item in itemslist1:
    name1 = item.find('div', 'view-field view-data-node-data-field-pet-name-field-pet-name-value').text.strip()
    info1 = item.find('div', 'view-field view-data-node-data-field-pet-look-field-pet-look-value').text.strip()
    l1.get('名字').append([name1])
    l1.get('簡介').append([info1])

df1 = pd.DataFrame(l1)
df1
```

```python
url2 = "https://wepet.tw/%E5%AF%B5%E7%89%A9%E9%A0%98%E9%A4%8A"
soup2 = BeautifulSoup((requests.get(url2)).text, "lxml")
itemslist2 = soup2.find_all('div', 'col-md-6')

p2 = []
for item in itemslist2:
    picture2 = item.find('img')
    p2.append(picture2)
for i in p2:
    if i == None:
        p2.remove(i)
p2
```

```python
l2 = {"名字":[],"簡介":[]}
for item in itemslist2:
    name2 = item.find('a','text-white text-capitalize d-block text-truncate')
    info2 = item.find('p')
    l2.get('名字').append(name2)
    l2.get('簡介').append(info2)

l2.get('名字').remove(None)
l2.get('簡介').remove(None)
df2 = pd.DataFrame(l2)
df2
```

```python
url3 = 'https://www.i-adopt.com.tw/adopt/cat'
soup3 = BeautifulSoup((requests.get(url3)).text, "lxml")
itemslist3 = soup3.find_all('div', 'adopt_unit hover_shadow')

p3 = []
for item in itemslist3:
    picture3 = item.find('img')
    p3.append(picture3)
p3
```

```python
l3 = {"名字":[],"簡介":[]}
for item in itemslist3:
    name3 = item.find_all('p','text-gray')
    info3 = item.find('p', 'text-gray desp').text.strip()
    l3.get('名字').append([name3[1].text.strip('動物名字:')])
    l3.get('簡介').append([info3])
df3 = pd.DataFrame(l3)
df3
```
